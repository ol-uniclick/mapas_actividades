# Mapas Python

```python
import sys
sys.path.append('./')
from mapas import *
```

Se genera el objeto de la clase `Mapa` indicando la dirección donde se encuentra el archivo csv.

```python
mapa = Mapa('mapa_demo.csv')
```

Para visualizar el mapa llamamos al atributo `df`.

```python
mapa.df
```

El método `fix_format` convierte todos los nombres y valores de las columnas a mayúsculas y elimina los acentos.

```python
mapa.fix_format()
```

En caso de que los valores de las columnas de ID se hayan cargado como números enteros, el siguiente método `fix_id` se puede usar para convertirlos al formato correcto, es decir:

\begin{equation}
1 \rightarrow "01"
\end{equation}

```python
mapa.fix_id()
```

El método `fix_dtypes` corrige los datatypes de todas las columnas, convierte todas las columnas a string, a excepcion de la columna "SEMANA" la cual convierte a número entero.

```python
mapa.fix_dtypes()
```

Cada semana es necesario actualizar la columna de "SEMANA" con la semana en curso, el método `update_week` realiza este trabajo.

```python
mapa.update_week()
```

Después de actualizar la columna "SEMANA" es necesario actualizar los valores de las columnas "SEMANA_INICIO" y "SEMANA_FIN". Para esto se utiliza el método `update_status`.

Este método únicamente actualiza las semanas que no están como RECURRENTE o TERMINADO.

```python
mapa.update_status()
```

Para poner un valor de la columna ESTATUS_PUNTUAL como 'TERMINADO' o 'RECURRENTE' se pueden usar los métodos `set_status`.

Si se aplica `mapa.set_status('TERMINADO', n)`, pone el valor de la columna ESTATUS_PUNTUAL con n-ésimo renglón como 'TERMINADO'. 


Análogamente para el caso 'RECURRENTE', se puede camiar el valor del n-ésimo renglón con:

Para regresar al valor de semanas que se calculó con el método `update_status` se usa el método `set_status` con el argumento `'SEMANAS'`. Pasándole como argumento el número de renglón donde se desea realizar este cambio.

Checamos el resultado final de aplicar todas las correciones.

```python
mapa.df
```

Finalmente para guardar el archivo se utiliza el método `to_file`. 

Si no se le pasa un argumento `filename` por default el nombre asignado será 'mapa_TIMESTAMP.txt', donde timestamp es una estampa de tiempo generada al momento de inicializar el objeto de clase `Mapa`. También se le puede indicar el directorio en donde se desea guardar el archivo a través del argumento `path`.

```python
mapa.to_file()
```

Es  opcional que realice las actualizaciones en las columnas de semanas y por default está como `update=True`. Si solo se quieren realizar las correciones de formato y no actualizar se pasa el argumento `update=False`.

```python
mapa2 = Mapa('mapa_demo.csv')
mapa2.df
```

```python
mapa2.fix_all()
```
